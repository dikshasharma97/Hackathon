```python
# defining RMSLE metric (Root mean squared log error)
```

```python
from sklearn.metrics import mean_squared_log_error
def rmsle(y, yhat):
    return (np.sqrt(mean_squared_log_error(y, yhat)))*1000

def rmsle_xg(yhat, y):
    y = np.expm1(y.get_label())
    yhat = np.expm1(yhat)
    return "rmsle", rmsle(y,yhat)
```

```python
# as the target variable is skewed applying log1p transformation
```

```python
V_xtrain = V_train.drop(['Sales','ID'],axis=1 )
V_ytrain = np.log1p(V_train.Sales)
V_xtest = V_test.drop(['Sales','ID'],axis=1 )
V_ytest = np.log1p(V_test.Sales)
```

# Model Building

```python

import xgboost as xgb
from time import time
```

```python


params = {"objective": "reg:linear",
          "booster" : "gbtree",
          "eta": 0.03,
          "max_depth": 14,
          "subsample": 0.9,
          "colsample_bytree": 0.7,
          "silent": 1,
          "seed": 10
          }
num_boost_round = 6000


dtrain = xgb.DMatrix(V_xtrain, V_ytrain)
dvalid = xgb.DMatrix(V_xtest, V_ytest)
watchlist = [(dtrain, 'train'), (dvalid, 'eval')]
```

```python
print("Train a XGBoost model")
start = time()
gbm = xgb.train(params, dtrain, num_boost_round, evals=watchlist, 
  early_stopping_rounds=100, feval=rmsle_xg, verbose_eval=True)
end = time()
print('Training time is {:2f} s.'.format(end-start))
```

```python
print("validating")
V_xtest.sort_index(inplace=True) 
V_ytest.sort_index(inplace=True) 
yhat = gbm.predict(xgb.DMatrix(V_xtest))
error = rmsle(np.expm1(V_ytest), np.expm1(yhat))

print('RMSPE: {:.6f}'.format(error))
```

```python
res = pd.DataFrame(data = V_ytest)
res['Prediction']=yhat
res = pd.merge(V_xtest,res, left_index= True, right_index=True)
res['Ratio'] = res.Prediction/res.Sales
res['Error'] =abs(res.Ratio-1)
res['Weight'] = res.Sales/res.Prediction
res.head()
```

```python
for i in range(1,5):
    s1 = pd.DataFrame(res[res['Course_ID']==i],columns = ['Sales','Prediction'])
    s2 = pd.DataFrame(res[res['Course_ID']==i],columns = ['Ratio'])
    s1.plot(figsize=(12,4))
    s2.plot(figsize=(12,4))
    plt.show()
```

```python
data_test2=data_test1.drop(['ID'],axis=1)
```

```python
print("Make predictions on the test set")

dtest = xgb.DMatrix(data_test2)
test_probs = gbm.predict(dtest)

# model1  kaggle private score 0.12647
result = pd.DataFrame({"ID": data_test1['ID'], 'Sales': np.expm1(test_probs)})
```

```python
result
```

```python
print("weight correction")
W=[(0.995+(i/1000)) for i in range(30)]
S =[]
for w in W:
    error = rmsle(np.expm1(V_ytest), np.expm1(yhat*w))
    print('RMSPE for {:.3f}:{:.6f}'.format(w,error))
    S.append(error)
Score = pd.Series(S,index=W)
Score.plot()
BS = Score[Score.values == Score.values.min()]
print ('Best weight for Score:{}'.format(BS))
```

# Hyperparameter Tuning

```python
metrics = pd.DataFrame(columns=['data_train','data_test'])
```

```python
import xgboost
from xgboost import XGBRegressor

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
estimator = XGBRegressor(subsample=0.8, colsample_bytree=1)
params = {'eta':[.3,.2,.1,.05,.01,.005], 'max_depth':[2,5,10,25], 'gamma':[0,0.01,0.1,1], 'min_child_weight':[1,5,10,25]}

rscv = RandomizedSearchCV(estimator, params, n_iter=10, n_jobs=-1, verbose=3, scoring='neg_mean_squared_log_error', cv=3)
rscv.fit(V_xtrain,V_ytrain)
```

```python
rscv.best_estimator_
```

```python
xgb = rscv.best_estimator_

xgb.fit(V_xtrain, V_ytrain)
#Train
y_pred_t = xgb.predict(V_xtrain)
train_score = rmsle(V_ytrain,y_pred_t)

#Test
y_pred_cv = xgb.predict(V_xtest)
test_score = rmsle(V_ytest,y_pred_cv)

metrics.loc['XGBRegressor'] = [train_score, test_score]

print(xgb.score(V_xtrain, V_ytrain), xgb.score(V_xtest, V_ytest))
```
